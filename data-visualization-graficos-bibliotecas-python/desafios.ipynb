```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/rafaelreisramos/oracle-one/refs/heads/main/data-visualization-graficos-bibliotecas-python/imigrantes_canada.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
imigracao = df.loc[['Brasil','Argentina'], anos].T
imigracao.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(imigracao['Brasil'], label='Brasil')
plt.plot(imigracao['Argentina'], label='Argentina')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.legend()
plt.show()
```
