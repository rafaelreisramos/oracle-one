## Dados

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/rafaelreisramos/oracle-one/refs/heads/main/data-visualization-graficos-bibliotecas-python/imigrantes_canada.csv'
df = pd.read_csv(url)
df.head()
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
df.head()
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
type(brasil)
```

```python
brasil_dict = {'ano': brasil.index.to_list(), 'imigrantes': brasil.values.tolist()}
brasil_dados = pd.DataFrame(brasil_dict)
brasil_dados.head()
```

## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(brasil_dados['ano'], brasil_dados['imigrantes'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(brasil_dados['ano'], brasil_dados['imigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de imigrantes')

plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(15,5))

axs[0].plot(brasil_dados['ano'], brasil_dados['imigrantes'])
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_title('Imigração do Brasil para o Canadá\n1980 a 2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numero de imigrantes')
axs[0].grid()

axs[1].boxplot(brasil_dados['imigrantes'])
axs[1].set_title('Boxplot da imigração do Brasil para o Canadá\n1980 a 2013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Numero de imigrantes')
axs[1].grid()

plt.show()
```

```python
brasil_dados.describe()
```

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Peru', anos])
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel('Ano')
    ax.set_ylabel('Número de imigrantes')

YMIN = 0
YMAX = 7000
# ravel() retorna um novo ax e não altera o limites do ax original
for ax in axs.ravel():
    ax.set_ylim(YMIN, YMAX)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(brasil_dados['ano'], brasil_dados['imigrantes'], lw=3, color='g')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigração do Brasil para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Número de imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

fig.savefig('imigracao_brasil_canada.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(america_sul.index, america_sul['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
america_sul_sorted = america_sul.sort_values('Total', ascending=True)
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown',
'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, ax = plt.subplots(figsize=(12,5))

ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('Imigração da América do Sul para o Canadá\n1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('g')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período entre 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

plt.show()
```

```python
cores = []

for pais in america_sul_sorted.index:
    if pais == 'Brasil':
        cores.append('g')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color=cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes\npara o Canadá no período entre 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
ax.tick_params(axis='both', which='both', length=0)
ax.get_xaxis().set_visible(False)

fig.savefig('imigracao_america_sul.png', transparent=False, dpi=300, bbox_inches='tight')

plt.show()
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.set_theme()
```

```python
top_10 = df.sort_values('Total', ascending=False).head(10)
top_10
```

```python
sns.barplot(data=top_10, x=top_10.index, y='Total')

plt.show()
```

```python
sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')

plt.show()
```

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set(title='Países com maior número de imigrantes para o Canadá\n1980 a 2013',
             xlabel='Número de imigrantes',
             ylabel='')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set_title('Países com maior número de imigrantes para o Canadá\n1980 a 2013', loc='left', fontsize=16)
ax.set_xlabel('Número de imigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```

```python
def gerar_grafico_paleta(palette):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h', palette=palette, hue=top_10.index, legend=False)
    ax.set_title('Países com maior número de imigrantes para o Canadá\n1980 a 2013', loc='left', fontsize=16)
    ax.set_xlabel('Número de imigrantes', fontsize=14)
    ax.set_ylabel('')

    plt.show()
```

```python
gerar_grafico_paleta('Blues')
```

```python
sns.set_theme(style='dark')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='white')
gerar_grafico_paleta('tab10')
```

```python
sns.set_theme(style='ticks')
gerar_grafico_paleta('tab10')
```
