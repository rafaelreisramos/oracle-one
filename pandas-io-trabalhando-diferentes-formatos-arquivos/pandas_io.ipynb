## Arquivos .json

```python
import pandas as pd
```

```python
url_1 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/pacientes.json'
```

```python
dados_pacientes = pd.read_json(url_1)
dados_pacientes.head()
```

Agora um .json com dados aninhados.

Para visualizar os arquivos .json você pode carregá-los no site https://jsoncrack.com/

```python
url_2 = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/pacientes_2.json'
```

```python
dados_pacientes_2 = pd.read_json(url_2)
dados_pacientes_2.head()
```

```python
dados_pacientes_normalizados = pd.json_normalize(dados_pacientes_2['Pacientes'])
dados_pacientes_normalizados.head()
```

Mas agora temos os dados normalizados mas somente para a lista pacientes e não todo o .json.

Para resolver isso podemos fazer o seguinte

```python
import requests

response = requests.get(url_2)
dados_pacientes_2_json = response.json()

dados_completos = pd.json_normalize(dados_pacientes_2_json, record_path=['Pacientes'], meta=['Pesquisa', 'Ano'])
dados_completos.head()
```

```python
dados_completos.to_json('historico_pacientes_normalizado.json')
```
