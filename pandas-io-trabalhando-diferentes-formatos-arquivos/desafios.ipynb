## Desafio 1

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/dados_sus.csv'
```

```python
dados = pd.read_csv(url)
```

```python
dados = pd.read_csv(url, skiprows=3, sep=';', encoding='ISO-8859-1')
dados.head()
```

```python
dados.tail()
```

```python
dados[:-7].tail()
```

Queremos eliminar também a coluna total

```python
dados = dados[:-8]
dados.tail()
```

Outra forma é usar `skipfooter` com `engine='python'`

```python
dados = pd.read_csv(url, skiprows=3, skipfooter=10, sep=';', encoding='ISO-8859-1', engine='python')
dados.tail()
```

## Desafio 3: normalizando o JSON obtido na API

```python
import pandas as pd
```

```python
url = 'https://jsonplaceholder.typicode.com/users'
```

```python
import requests

response = requests.get(url)
users = response.json()
```

```python
import json

print(json.dumps(users, indent=4))
```

```python
users_normalizado = pd.json_normalize(users, sep='_')
users_normalizado.head()
```
